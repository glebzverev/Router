```python
import requests
from pprint import pprint
```

```python
def run_query(q):

    # endpoint where you are making the request
    request = requests.post('https://api.thegraph.com/subgraphs/name/aave/protocol'
                            '',
                            json={'query': query})
    if request.status_code == 200:
        return request.json()
    else:
        raise Exception('Query failed. return code is {}.      {}'.format(request.status_code, query))


# The Graph query - Query aave for a list of the last 10 flash loans by time stamp
query = """

{
flashLoans (first: 10, orderBy: timestamp, orderDirection: desc,){
  id
  reserve {
    name
    symbol
  }
  amount
  timestamp
}
}
"""
result = run_query(query)

# print the results
print('Print Result - {}'.format(result))
print('#############')
# pretty print the results to make it easier to read
pprint(result)
```

```python
columns = ['id', 'token0_id', 'token0_name', 'reserve0', 'token0Price', 'volumeToken0', 'token1_id', 'token1_name', 'reserve1', 'token1Price','volumeToken1']
```

```python
import pandas as pd

pairs = pd.DataFrame(columns = columns)
```

```python
from tqdm import tqdm
API_KEY = "66b80d3d689772ba05ed3461d4d6cb30"
endpoint = "https://gateway.thegraph.com/api/"

def concatURI(index):
    return "{}{}/subgraphs/id/{}".format(endpoint,API_KEY,index)

def ammQuery(amount, skip_amount):
    return "{pairs(first: " + str(amount) + ", skip: " + str(skip_amount) + ") { id token0{ id name }     reserve0 token0Price volumeToken0 token1{ id name } reserve1 token1Price volumeToken1 } }";

query_shape = 10

for j in tqdm(range(10)):
    for i in subgraph['AMM']:
        request  = requests.post(
            concatURI(subgraph['AMM'][i]), 
            json = {"query": ammQuery(query_shape, j*query_shape)}
        )
        if request.status_code == 200:
            for i in request.json()['data']['pairs']:
                item = i
                item['token0_id'] = i['token0']['id']
                item['token0_name'] = i['token0']['name']
                item['token1_id'] = i['token1']['id']
                item['token1_name'] = i['token1']['name']
                item.pop('token0')
                item.pop('token1')
                pairs = pairs.append(i, ignore_index=True)
        else:
            print( Exception('Query failed. return code is {}.      {}'.format(request.status_code, query)))
```

```python
pairs.shape, len(set([i for i in pairs.token0_name.unique()]).union(pairs.token1_name.unique()))
```

```python
transactions = [[pairs.iloc[i]['token0_name'], pairs.iloc[i]['token1_name']] for i in tqdm(range(pairs.shape[0]))]
```

```python
import networkx as nx
import matplotlib.pyplot as plt

G=nx.MultiDiGraph()
for t in transactions:
    for i in range(len(t) - 1):
        G.add_edge(t[i],t[i+1])
        
plt.figure(figsize=(20, 14))
pos = nx.spring_layout(G)
nodes = nx.draw_networkx_nodes(G, pos)
edges = nx.draw_networkx_edges(G, pos, alpha=0.01)
labels = nx.draw_networkx_labels(G, pos)
```
